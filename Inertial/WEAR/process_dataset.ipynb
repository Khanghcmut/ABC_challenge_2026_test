```python
import pandas as pd
import os
```

```python
df=pd.read_csv('./Raw_data/sbj_0.csv')
df
```

```python
def to_dataframe(df:pd.DataFrame,modality:str):
    df_out=df.loc[:,df.columns.str.contains('|'.join([modality,'label']))]
    df_out.columns=['_'.join(i.split('_')[1:]) for i in df_out.columns]
    df_out.columns=df_out.columns[:-1].to_list()+['label']
    return df_out
df_=to_dataframe(df,'left')
df_
```

```python
os.makedirs('Processed_data',exist_ok=True)
for i in os.listdir('./Raw_data'):
    df=pd.read_csv(os.path.join('./Raw_data',i))
    df_left=to_dataframe(df,'left')
    df_right=to_dataframe(df,'right')
    df_left.to_csv(os.path.join('Processed_data',f'1_{i}'),index=False)
    df_right.to_csv(os.path.join('Processed_data',f'2_{i}'),index=False)
    
```

```python
os.makedirs('./Processed_data_shrink',exist_ok=True)
for i in os.listdir('./Processed_data'):
    if i.endswith('.csv'):
        df=pd.read_csv(f'./Processed_data/{i}')
        df_shrink=df.iloc[::2,:]
        df_shrink.to_csv(os.path.join('./Processed_data_shrink',i),index=False)
```
